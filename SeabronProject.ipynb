```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('hotel_bookings.csv')
print(df.head(10))
print(df.info())
print(df.describe())
```

```python
#Distribution of Lead Time
sns.set(style='whitegrid')
sns.histplot(df['lead_time'],kde=True,bins=30)
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
#2-Distrubution of Cancellations
sns.set(style='whitegrid')
plt.figure(figsize=(8,6))
sns.countplot(x='is_canceled',data=df,palette='viridis')
plt.title('Distribution of Cancellations')
plt.xlabel('Is Cancelled')
plt.ylabel('Count')
plt.xticks([0,1],['Not Cancelled','Canceled'])
plt.show()
```

```python
#3- Monthly Distribution of Cancellations
sns.countplot(x='arrival_date_month',hue='is_canceled',data=df,palette='Accent')
plt.title('Monthly Distribution of Cancellations')
plt.xlabel('Arrival Month')
plt.ylabel('Count')
plt.legend(title='Is Cancelled',labels=['Not Canceled','Canceled'])
plt.xticks(rotation=90)
plt.show()
```

```python
#4- Distribution of ADR
plt.figure(figsize=(10,6))
sns.barplot(x='total_of_special_requests',y='is_canceled',data=df,palette='cool')
plt.title('Number of special Requests by Reservation Status')
plt.xlabel('Total Special Requests')
plt.ylabel('Cancelation Rate')
plt.show()
```

```python
#5- Number of Special Requests by Reservation Status
plt.figure(figsize=(10,6))
sns.barplot(x='total_of_special_requests',y='is_canceled',data=df,palette='tab10')
plt.title('Number of Special Requests by Reservation Status')
plt.xlabel('Total Special Requests')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
#6- Customer Type vs Cancellation
plt.figure(figsize=(10,6))
sns.countplot(x='customer_type',hue='is_canceled',data=df,palette='Set2')
plt.title('Customer Type vs Cancellation')
plt.xlabel('Customer Type')
plt.ylabel('Count')
plt.legend(title='Is Canceled',labels=['Not Canceled','Canceled'])
plt.show
```

```python
#7- Market Segment vs ADR
plt.figure(figsize=(12, 6))
sns.boxplot(x='market_segment', y='adr', data=df, palette='viridis')
plt.title('Market Segment vs. ADR')
plt.xlabel('Market Segment')
plt.ylabel('ADR')
plt.xticks(rotation=45)
plt.show()
```

```python
#8- Heatmap of correlations
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Display the selected numeric columns to verify
print(numeric_df.head())

# Create a heatmap of correlations between numerical features
plt.figure(figsize=(14, 10))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Heatmap of Correlations')
plt.show()
```
